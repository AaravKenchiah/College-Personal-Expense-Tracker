```python
''' Average Food Spending Function '''
def analyzing_food_spending(df):
    average_food = df['food'].mean()
    print(f'Average student spends ${average_food} on food')
    min_food = df['food'].min()
    max_food = df['food'].max()
    percentile_25 = df['food'].quantile(.25)
    percentile_75 = df['food'].quantile(.75)
    return
```

```python
''' Food Spennding by Year ''' 
def food_by_year(df):
    print(df['year_in_school'].value_counts())
    grade_sort = df.groupby('year_in_school')
    avg_food_cost_grade = grade_sort['food'].mean()
    print(f"\nAverage food spending by grade: \n{avg_food_cost_grade}")
    return
```

```python
''' food vs income '''
def foods_vs_income(df):
    df['food_percentage'] = (df['food']/df['monthly_income']) * 100
    print('Food spending as percent of income')  
    print(df['food_percentage'].describe())
    income_ranges = df.groupby(pd.cut(df['monthly_income'], bins = 3))['food'].mean()
    print(income_ranges)
    return
```

```python
def food_budget_reccs(df):
    try:
        income = float(input("Please enter your monthly income: "))
        if income > df["monthly_income"].max():
            print("Warning: Your income is higher than most students in our dataset, so recommendations may be less accurate")
        

    except ValueError:
        print("Please enter a valid number for income.")
        return

    df['income_distance'] = abs(df['monthly_income'] - income) #creating a 'range' of how far inputed budget is from already selected budgets
    df_sorted = df.sort_values('income_distance')
    similar_students = df_sorted.head(50)

    avg_food = similar_students['food'].mean()
    min_food = similar_students['food'].min()
    max_food = similar_students['food'].max()
    
    print(f"Based on {len(similar_students)} students with similar income:")
    print(f"Average food spending: ${avg_food:.2f}")
    print(f"Range: ${min_food:.2f} - ${max_food:.2f}")
    return
            

```

```python
''' RUN ALL FUNCTIONS '''

print("🎓 STUDENT EXPENSE ANALYSIS")
print("=" * 50)

print("\n1️⃣ BASIC FOOD SPENDING STATS:")
print("-" * 30)
analyzing_food_spending(df)

print("\n2️⃣ FOOD SPENDING BY YEAR:")
print("-" * 30)
food_by_year(df)

print("\n3️⃣ FOOD VS INCOME ANALYSIS:")
print("-" * 30)
foods_vs_income(df)

print("\n4️⃣ PERSONALIZED BUDGET RECOMMENDATIONS:")
print("-" * 30)
food_budget_reccs(df)

print("\n✅ All analysis complete!")
```
